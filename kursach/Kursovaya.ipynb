Преобразую этот текстовый признак в числовой (количество месяцев)

```python
import re

def convert_credit_history_age(text):
    if pd.isna(text) or text == '':
        return None
    years_match = re.search(r'(\d+)\s*Years?', text)
    months_match = re.search(r'(\d+)\s*Months?', text)
    
    years = int(years_match.group(1)) if years_match else 0
    months = int(months_match.group(1)) if months_match else 0
    total_months = years * 12 + months
    return total_months
```

```python
df['Credit_History_Age'] = df['Credit_History_Age'].apply(convert_credit_history_age)
```

Нужно очистить признак

```python
def clean(value):
    if pd.isna(value):
        return np.nan
    
    as_str = str(value)
    digits = ''.join([ch for ch in as_str if ch.isdigit()])
    
    if digits == '':
        return np.nan
    
    num = int(digits)
    
    if '-' in as_str and num > 0:
        num = abs(num)
    
    return num
```

```python
df['Num_of_Delayed_Payment'] = df['Num_of_Delayed_Payment'].apply(clean)
```

## Построение моделей кредитного скоринга

```python
X = df.drop('Credit_Score', axis=1)
y = df['Credit_Score']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=245, stratify=y)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### Обучение логистической регрессии (One-vs-Rest)

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr_model = LogisticRegression(multi_class='ovr', class_weight='balanced', max_iter=1000, random_state=245)
```

```python
lr_model.fit(X_train_scaled, y_train)
```

```python
y_pred_lr = lr_model.predict(X_test_scaled)
```

###  Обучение LinearSVC

```python
from sklearn.svm import LinearSVC
```

```python
svc_model = LinearSVC(class_weight='balanced', C=1.0, max_iter=2000, random_state=245, dual=False, penalty='l2', loss='squared_hinge')
```

```python
svc_model.fit(X_train_scaled, y_train)
```

```python
y_pred_svc = svc_model.predict(X_test_scaled)
```

### Обучение случайного леса

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=245, n_jobs=-1)
```

```python
rf_model.fit(X_train, y_train)
```

```python
y_pred_rf = rf_model.predict(X_test)
```

### Обучение XGBoost

```python
import xgboost as xgb
```

```python
xgb_model = xgb.XGBClassifier(
    n_estimators=100,
    max_depth=6,
    learning_rate=0.1,
    objective='multi:softmax',      
    num_class=3,
    random_state=245,
    eval_metric='mlogloss',
    use_label_encoder=False
)
```

```python
xgb_model.fit(X_train, y_train)
```

```python
y_pred_xgb = xgb_model.predict(X_test)
```

### Обучение CatBoost

```python
from catboost import CatBoostClassifier
```

```python
cat_model = CatBoostClassifier(
    iterations=100,
    depth=6,
    learning_rate=0.1,
    random_seed=245,
    verbose=False
)
```

```python
cat_model.fit(X_train, y_train)
```

```python
y_pred_cat = cat_model.predict(X_test)
```
